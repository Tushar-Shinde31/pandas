groupby() method -> to split the data into groups and then apply any operation on groups 

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\pandas\adv-datasets\cities_weather.csv')
df
```

```python
df1 = df.groupby('city')
```

```python
for city, city_df in df1:
    print(city_df)
```

```python
df1.get_group('mumbai')
```

Aggregate function : are the function that act on group of element
    max(), min(), mean(), sum(), median(), std(), prod(), count()

```python
df2 = df1.get_group('mumbai')
# df1.get_group('mumbai').max()
```

```python
df2.agg('max')
```

```python
df2['temperature'].agg(['max', 'min'])
```

```python
# find avg speed in New york
df1 = df.groupby('city')['windspeed'].mean()
df1
```

```python
df2 = df1.get_group('new york')
print('Avg new york windspeed: ',df2['windspeed'].agg('mean'))
```

```python
# min temp and max windspeed in new york
df2 = df1.get_group('new york')
print('minimum in tempreature in new york: ',df2['temperature'].agg('min'))
print('maximum in tempreature in new york: ',df2['windspeed'].agg('max'))
```

```python
# 2nd way
df2.agg({'temperature' : 'min', 'windspeed': 'max'})
```

```python
#  find min, max on temperature - sum and avg on windspeed in new york
df2.agg({'temperature' : ['min', 'max'], 'windspeed' : ['sum', 'mean']})
```

```python
df[df['city'] == 'new york'].agg({
    'temperature': ['min', 'max'],
    'windspeed': ['sum', 'mean']
})
```

```python
# its for total dataset
df.agg({'temperature' : ['min', 'max'], 'windspeed' : ['sum', 'mean']})
```
