```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'a': [1, 2, 34, 5, 6, 7, 8],
    'b': [10, np.nan, np.nan, 20, np.nan, np.nan, 30]
})

print(df)
```

```python
# find cummulative sum of the col 'a'
df['a_cumsum'] = df['a'].cumsum()
df
```

```python
# cummu prod of 'a'
df['a_cumsum'] = df['a'].cumprod()
df
```
