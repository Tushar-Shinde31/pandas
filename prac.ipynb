```python
text = "My age is 22 and my brother is 18"

import re

res = re.findall(r'\d', text)

print(res)
```

```python
res1 = re.findall(r'\d+', text)

print(res1)
```

```python
res1 = re.findall(r'\D', text)

print(res1)
```

```python
text = 'Python     Java SQL'

res = re.findall(r'\S', text)

print(res)
```

```python
text = 'Python123'

if re.match(r'\d+', text):
    print('start with digit')
else:
    print('no')
```

```python
text = """
abc@gmail.com
xyz@yahoo.com
test@outlook.com
"""

emails = re.findall(r'\w+@\w+\.\w+', text)

print(emails)

```

```python
text = "cat dog elephant bat sun"

res = re.findall(r'\b\w{3}\b', text)

print(res)
```

```python
text = "My age is 22"

res = re.sub(r'\d', 'X', text)

print(res)
```

```python
import re

text = "color colour column"

print(re.findall(r"colou?r", text))
```

```python
text = "I was born in 2001 and graduated in 2024"

res = re.findall(r'\d{4}', text)

print(res)
```

```python
text = "Data Engineering Roadmap"

res = re.sub(r'\s', '-', text)

print(res)
```

```python
text = "Tushar works at Google in Pune"

res = re.findall(r'\b[A-Z]\w*', text)

print(res)

```

```python
text = "apple mango cat house tiger"

res = re.findall(r'\b\w{5}\b', text)

print(res)
```

```python
import re

data = """
Tushar,22,Python
Rohit,25,SQL
Amit,28,Spark
"""

records = re.findall(r"(\w+),(\d+),(\w+)", data)

print(records)
```

```python
log = """
ERROR: Disk Full
INFO: Job Started
ERROR: Memory Low
INFO: Job Completed
"""


res = re.findall(r'ERROR:.*', log)

print(res)
```
