window -> one or more col of a df,
window func -> time series

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'a': [1, 2, 34, 5, 6, 7, 8],
    'b': [10, np.nan, np.nan, 20, np.nan, np.nan, 30]
})

print(df)
```

```python
# find cummulative sum of the col 'a'
df['a_cumsum'] = df['a'].cumsum()
df
```

```python
# cummu prod of 'a'
df['a_cumprod'] = df['a'].cumprod()
df
```

Rolling window: means moving window

```python
# rolling window to find rolling sum of 'a'
obj = df['a'].rolling(window = 2)
df['a_roll_sum'] = obj.sum()
df
```

```python
# find rolling count on col b
obj = df['b'].rolling(window = 2, min_periods=2)
df['b_roll_cnt'] = obj.count()
df
```

cummulative nd expandind window are same

```python
# find mean val on expanding window
obj = df['a'].expanding(min_periods=1)
df['a_exp_mean'] = obj.mean()
df
```
