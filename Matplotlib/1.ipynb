```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\pandas\Dataset\score.csv')
df
```

Line Plot

```python
# line plot
x = df['marks']
y = df['rank']
plt.plot(x,y, color='orange')
plt.xlabel('Student marks', color='red')
plt.ylabel('Rank', color='blue')
plt.title('EXAMS MARKS RANK')
plt.show
```

Scatter Plot

```python
#  scatter plot
x = df['marks']
y = df['rank']
plt.plot(x,y, color='orange')
plt.scatter(x,y, color='green')
plt.xlabel('Student marks', color='red')
plt.ylabel('Rank', color='blue')
plt.title('EXAMS MARKS RANK')
plt.show
```

Bar Graph

```python
# bar graph
x = df['marks']
y = df['rank']
plt.bar(x,y, color='orange', label='bar graph')
plt.xlabel('Student marks', color='red')
plt.ylabel('Rank', color='blue')
plt.title('EXAMS MARKS RANK')
plt.show()
```

Pie Chart

```python
x = df['marks']
```

```python
#  how many students got above 850 marks
x1 = len(x[x>850])
x1
```

```python
# how many got marks between 800 and 900
x2 = len(x[(x>800) & (x<900)])
x2
```

```python
# how many student got marks less than 800
x3 = len(x[x<800])
x3
```

```python
print(x1, x2, x3)
```

```python
slices = [x1, x2, x3]
cols = ['yellow', 'red', 'green']
lbls = ['>900', '800 to 900', '<800']

plt.pie(
    slices,
    labels=lbls,
    colors=cols,
    startangle=90,
    shadow=True,
    explode=(0.2, 0, 0),
    autopct= '%.1f%%'
)

plt.show()
```
