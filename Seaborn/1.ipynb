```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
arr = mpg['cylinders'].unique()
tpl = tuple(mpg['cylinders'])
```

Displot

```python
sns.displot(tpl, bins=6)
```

```python
sns.displot(tpl, kde=True, bins=6)
```

```python
sns.displot(tpl, kde=False, bins=range(3, 10, 1))
```

```python
mpg['origin'].unique()
```

Count Plot


```python
sns.countplot(x='cylinders', data=mpg, color='red', hue='origin')
```

Box Plot

```python
sns.boxplot(x='origin', y='acceleration', data=mpg)
```

Joint plot

```python
sns.jointplot(x='horsepower', y='acceleration', data=mpg)
```

Scatter Plot

```python
sns.scatterplot(x = 'horsepower', y='acceleration', data=mpg, alpha = 0.2)
```

```python
sns.scatterplot(x='horsepower', y='acceleration', data=mpg, hue='origin')
```

Sub plots

```python
plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
sns.scatterplot(x='acceleration', y='horsepower', data=mpg)

plt.subplot(2,2,2)
sns.scatterplot(x='acceleration', y='origin', data=mpg)

plt.subplot(2,2,3)
sns.boxplot(x='acceleration', y='origin', data=mpg)
```

Lmplot

```python
sns.lmplot(x='horsepower', y='acceleration', data=mpg, hue='origin')
```
