```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
print(len(iris))
```

Pair plot

```python
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, hue='species', palette='husl', markers=['o', 'd', 's'])
```

```python
sns.pairplot(iris, vars=['sepal_length', 'sepal_width'])
```

Heatmap

```python
df = iris.iloc[:10, :4]
df
```

```python
sns.heatmap(df, annot=True, cmap='BrBG')
```

Cat plot

```python
sns.catplot(x='species', y='sepal_width', data=iris, kind='violin')
```
