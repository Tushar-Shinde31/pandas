```python
import pandas as pd
```

```python
tpl = ((1001, 'Tushar'), (1002, 'Satyarth'), (1003, 'Rohit'), (1004, 'Shivam'))
```

```python
df = pd.DataFrame(tpl, columns=['col1', 'col2'])
df
```

```python
#  rename the col
df.rename(columns = {'col1' : 'empid', 'col2' : 'name'}, inplace=True)
df
```

```python
# add third col or new col
df['dept'] = ['IT', 'HR', 'Finance', 'Admin']
df
```

```python
#  how to insert new col in betw existing col
df.insert(loc=1, column = 'salary', value = [35000, 22000, 50000, 54001])
```

```python
df
```

```python
#  iosert phone number col in betw name and dept col
df.insert(loc=3, column = 'phone', value = [1234567890, 9876543210, 4567891230, 7891234560])
df
```

```python
#  create a new col from existing col
df['Tax'] = df['salary'] * 10/100
df
```

```python
#  create a new col from exiting col using lambda
df1 = df.assign( PF = lambda x : x.salary * 12.5/100)
df1
```

```python
# cal net salary net sal = salary - tax - pf
df2 = df1.assign(net_salary = lambda x : x.salary - x.Tax - x.PF)
df2
```

```python
#  cal deduction = tax + pf
df3 = df2.assign(deductions = lambda x : x.Tax + x.PF)
df3
```
