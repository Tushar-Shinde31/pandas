Where method

```python
import numpy as np
```

```python
df = pd.DataFrame(data = np.random.randint(0, 100, (4,3)),
columns = ('Test1', 'Test2', 'Test3'),
index = ['Vinay', 'Tushar', 'Ajay', 'Kartik'])


df
```

```python
# if marks are greater than 70 or 80 replace those marks with A+ grade
df.where(df <= 70, 'A+')
```

```python
df.where(df['Test3'] >= 30, 'Bad')
```

```python
# where narks = 40 diplay medium marks
df.where(df != 40, 'Medium')
```

```python
#  using lambda
df.where(lambda x : x != 26, 'Medium')
```

```python
df1 = pd.DataFrame(data = np.random.randint(0, 100, (4,3)),
columns = ('Test1', 'Test2', 'Test3'),
index = ['Vinay', 'Tushar', 'Ajay', 'Kartik'])


df1
```

```python
# replacing the val based on cond from df to df1
df1.where(df <= 80, df1)
```

```python
df1.where(df <= 80, df + df1)
```
