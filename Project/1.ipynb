```python
# Write a Python program to find out the alcohol consumption details in the year '1987' or
# '1989' from the dataset
result = df[(df['Year'] == 1987) | (df['Year'] == 1989)]
result
```

```python
# Write a python program to find out and display horizontal bar chart the alcohol
# consumption details in the year '1987' where WHO region is 'Western Pacific' and country
# is 'VietNam' from thedataset.
res = df[(df['Year'] == 1987) & (df['WHO region'] == 'Western Pacific') & (df['Country'] == 'Viet Nam')]
res = res.dropna()
res
```

```python
import matplotlib.pyplot as plt
plt.barh(' Beverage Types', 'Display Value', data=res, color='orange')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Beverage Types')
plt.title('Alcohol Consumption in Viet Nam (1987)')
plt.show()
```

```python
# Write a Pandas program to filter those records where WHO region matches with multiple
# values (Africa, Eastern Mediterranean, Europe) from dataset.
result = df[df['WHO region'].isin(['Africa', 'Eastern Mediterranean', 'Europe'])]
result
```

```python
# Write a Pandas program to filter those records which not appears in a given list from
# dataset.
# who_region = ["Africa", "Eastern Mediterranean", "Europe"]
result = df[df['WHO region'].isin(['Africa', 'Eastern Mediterranean', 'Europe']) == False]
result
```

```python
# Write a Pandas program to find average consumption of wine per person greater than
# 2 in world alcohol consumption dataset.
result = df[(df['Beverage Types'] == 'Wine') & (df['Display Value'] > 2)]
result
```

```python
# Write a Pandas program to filter rows, based on row numbers ended with 0, like 0, 10,
# 20, 30 from dataset.
result = df[df.index % 10 == 0]
result
```

```python
# Write a Pandas program to also select rows with Index label 0 to 9 with some columns
# from dataset.
result = df.loc[0:9, ['WHO region', 'Country', 'Beverage Types']]
result
```

```python
# Write a Pandas program to rename all and only some of the column names from
# dataset by adding _ in column names.
result = df.rename(columns={'WHO region': 'WHO_region', 'Beverage Types': 'Beverage_Types', 'Display Value': 'Display_Value'})
result
```

```python
# Write a Pandas program to find which years have all non-zero values and which years
# have any non-zero values from dataset.
result = df[~df['Year'].isin([0])]
result
```

```python
# Write a Pandas program to filter all columns where all entries present, check which
# rows and columns has a NaN and finally drop rows with any NaNs from world alcohol
# consumption dataset.
result = df.dropna()
result
```

```python
# Write a Pandas program to filter all records starting from the 'Year' column, access
# every other column from world alcohol consumption dataset.
result = df.iloc[:, 1::2]
result
```

```python
# Write a Pandas program to filter all records starting from the 2nd row, access every 5th
# row from dataset.
result = df.iloc[2::5, :]
result
```

```python
# Write a Pandas program to create bar plots with error bars to filter all records where
# the average consumption of beverages per person from 0.5 to 2.50 in dataset.
result = df[(df['Display Value'] >= 0.5) & (df['Display Value'] <= 2.5)]
result
```

```python
# bar plot with error bars
plt.bar('Beverage Types', 'Display Value', data=result, color='green', yerr=0.5)
plt.xlabel('Beverage Types')
plt.ylabel('Average Consumption')
plt.title('Average Consumption of Beverages (0.5 to 2.5 per person)')
plt.show()
```
